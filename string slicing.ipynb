String format method: versatile method to embed the variable and format values within the string.

```python
print('This is my string {}'.format('InSERTED'))
```

```python
print('my name is {} and I am {} years old'.format('AMrita','30'))
```

keyword argument
You can name the placeholders within the curly braces and
then pass keyword arguments to format() with matching names.

```python
product='laptop'
price=1200
info='the {item} cost ${cost}'.format(item=product,cost=price)
print(info)
```

Indexing Arguments:
You can specify the index of the argument within the curly braces 
to control which argument replaces which placeholder.

```python
arguments=["12", "22", "33"]
print('the {} one to be inserted, followed by {} and the {}'.format(arguments[0],arguments[1],arguments[2]))
```

Format Specifiers
Within the curly braces, you can include format specifiers afer a colon:
to control how the values are presented(alignment, precision and number formatting)

```python
pi=3.14159
print ('pi is two decimal places: {:.2f}'.format(pi))
```
